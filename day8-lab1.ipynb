## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x, y,alpha = 0.4, marker = 'o', s = 50)
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x, y,alpha = 0.4, marker = 'o', s = 16)
plt.title('chart', fontsize = 16, loc='left', color='r');
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,0,4,5])
plt.plot([7,9,8,6,3])
plt.plot([10,2,7,3,8])
plt.legend(['1','2','3'], shadow = True, loc = 1)
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.subplot(1,3,1)
plt.plot([1,2,0,4,5])
plt.plot([7,9,8,6,3])
plt.plot([10,2,7,3,8])
plt.legend(['1','2','3'], shadow = True, loc = 1)

plt.subplot(1,3,2)
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x, y,alpha = 0.4, marker = 'o', s = 16)
plt.title('chart', fontsize = 16, loc='left', color='r');

plt.subplot(1,3,3)
plt.pie(np.arange(1,5));
```
